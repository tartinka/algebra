```python
1. Задание
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.
```

```python
def v_length(vektor):
    s = 0
    for a in vektor:
        s = a**2 + s
    return math.sqrt(s)

my_vekt = (4,3,2)
print(f'длина вектора {my_vekt} равна {v_length(my_vekt)}')

```

2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? 
Ответ: потому что масштаб осей разный. Если сделать одинаковым, то будет перпендикуляр.

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
rcParams["figure.figsize"] = 5, 5
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.axis([-15, 15, -15, 15])
plt.show()
```

3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы

```python
u=0    # центр по оси х
v=0   # центр по оси y
a=2     # радиус по оси x
b=2   # радиус по оси y

t = np.linspace(0, 2*np.pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
plt.axis('scaled')
plt.show()
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
print(X)
X, Y = np.meshgrid(X, Y)
print(Y)
print(X[1,1])
Z = 2*X + 3*Y + 50
Z2 = 2*X + 3*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
print(X)
X, Y = np.meshgrid(X, Y)
print(Y)
print(X[1,1])
Z = 2*X + Y**3 + 50
Z2 = X**2 + 3*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
